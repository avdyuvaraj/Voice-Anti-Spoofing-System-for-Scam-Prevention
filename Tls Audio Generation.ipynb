```python
This project aims to detect synthetic (AI-generated) voice samples from real human voice recordings
using machine learning techniques. We extract MFCC features and train a Random Forest classifier.
The real speech samples were obtained from publicly available speech datasets and controlled recordings.
Synthetic samples were generated using open-source text-to-speech systems to simulate voice spoofing scenarios.
The dataset was manually labeled into two categories: real human speech and synthetic speech generated
using open-source TTS systems.
```

```python
from gtts import gTTS

sentences = [
"This is a sample voice recording for spoof detection testing.",
"Cybersecurity systems must detect artificial speech accurately.",
"Voice authentication plays a critical role in fraud prevention.",
"Synthetic voices are becoming more realistic each year.",
"This project analyzes acoustic features of human speech.",
"Machine learning can identify subtle audio patterns.",
"Audio spoofing attacks are a growing concern worldwide.",
"We are testing a speech classification model today.",
"Real and fake voices often sound very similar.",
"Detecting voice manipulation is important for security systems.",
"This sentence is generated using text to speech technology.",
"The model extracts mel frequency cepstral coefficients.",
"Advanced fraud detection requires strong signal analysis.",
"Modern attackers use artificial intelligence for scams.",
"Voice biometrics can enhance authentication mechanisms.",
"This experiment evaluates a binary audio classifier.",
"Speech processing involves analyzing sound frequencies.",
"Secure communication depends on accurate identity verification.",
"Audio samples are processed at sixteen kilohertz.",
"The classifier distinguishes between real and synthetic speech.",
"Acoustic feature extraction improves detection performance.",
"Fraud prevention technologies must evolve continuously.",
"Voice spoofing detection is an active research area.",
"This recording simulates an artificial voice sample.",
"Security systems should detect manipulated audio effectively."
]

for i, sentence in enumerate(sentences):
    tts = gTTS(sentence)
    filename = f"fake_{i+1}.wav"
    tts.save(filename)

print("25 synthetic audio files generated successfully.")
```
